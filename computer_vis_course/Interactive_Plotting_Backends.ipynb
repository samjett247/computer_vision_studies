## Interactive Point Inputs using ginput

```python
import matplotlib 
matplotlib.use('macosx')
from matplotlib import pyplot as plt
import numpy as np


plt.figure()
x,y = np.random.rand(40), np.random.rand(40)
plt.plot(x,y)
plt.ginput(n=3)
plt.close()
print('yay')
```

```python
%matplotlib inline
plt.figure()
x,y = np.random.rand(40), np.random.rand(40)
plt.plot(x,y,'r:')
plt.show()
```

```python
%matplotlib osx
plt.figure()
x,y = np.random.rand(40), np.random.rand(40)
plt.plot(x,y)
points = plt.ginput(n=3)
plt.close()
print('yay')
print(*points)
# COOL!
```
